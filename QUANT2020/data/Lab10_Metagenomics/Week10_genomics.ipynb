```python
def bin_species_id(bin_num, assembly):
    f = open("/Users/cmdb/qbb2020-answers/QUANT2020/data/Lab10_Metagenomics/bins/bin."+bin_num+".fa")
    binfile = f.readlines()
    f.close()
    bin_dict = {}
    for line in binfile:
        if '>' in line:
            node = str(line.strip()[1:])
            for line in assembly:
                if node in line:
                    key = line.split('\t')[1]
                    bin_dict.setdefault(key,0)
                    bin_dict[key] += 1
                    
    f = open("/Users/cmdb/qbb2020-answers/QUANT2020/data/Lab10_Metagenomics/bins/bin_species."+bin_num+".fa",'w')
    for key,val in bin_dict.items():
        f.write(key + '\n')
        f.write(str(val)+'\n')
        f.write('\n')
    max_key = max(bin_dict, key=lambda k: bin_dict[k])
    f.write('\nMost likely associated species:\t'+max_key)
    f.close()
    
```
